# Axes labels, Legend, Grid

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,46,49]
min_t = [46,48,48,40,42,41,43]
avg_t = [48,49.5,50,44,44.5,43.5,46]
```

```python
plt.plot(days,max_t)
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,max_t,label='MAX')
plt.plot(days,min_t,label='MIN')
plt.plot(days,avg_t,label='AVG')

plt.legend(loc='lower left',shadow='True',fontsize='small')

plt.grid()
```

# Barchart

```python
Student_Name=['Nakib','Azrin','Chaity','Zayed','Dany','Mouri']
Marks=[93,82,74,80,61,54]
Viva=[18,14,16,7,12,10]
```

```python
plt.bar(Student_Name,Marks)
```

```python
y_pos=np.arange(len(Student_Name))
y_pos
```

```python
plt.ylabel('Marks')
plt.title('Result')

plt.xticks(y_pos,Student_Name)
plt.bar(y_pos,Marks)
```

```python
plt.ylabel('Marks')
plt.title('Result')

plt.xticks(y_pos,Student_Name)
plt.bar(y_pos,Marks,label='Marks',color='yellow')
plt.bar(y_pos,Viva,label='Viva',color='black')

plt.legend()
```

```python
plt.ylabel('Marks')
plt.title('Result')

plt.xticks(y_pos,Student_Name)
plt.bar(y_pos,Marks,width=0.5,label='Marks',color='yellow')
plt.bar(y_pos+0.2,Viva,width=0.5,label='Viva',color='black')

plt.legend()
```

```python
plt.xlabel('Marks')
plt.title('Result')

plt.yticks(y_pos,Student_Name)
plt.barh(y_pos,Marks,label='Marks',color='yellow')
plt.barh(y_pos-0.2,Viva,label='Viva',color='black')

plt.legend()
```

# Histogram

```python
Blood_suger=[112,103,95,89,115,128,85,150,138,99,140,122,111]
Blood_suger_women=[82,100,92,88,105,98,80,115,138,90,104,102,100]
plt.hist(Blood_suger)
```

```python
plt.hist(Blood_suger,bins=20)
```

```python
plt.hist(Blood_suger,bins=3,rwidth=0.95)
```

```python
plt.hist(Blood_suger,bins=[80,100,125,150],rwidth=0.95)
```

```python
plt.hist(Blood_suger,bins=[80,100,125,150],rwidth=0.9,color='g')
```

```python
plt.hist(Blood_suger,bins=[80,100,125,150],rwidth=0.95,histtype='step')
```

```python
plt.hist([Blood_suger,Blood_suger_women],bins=[80,100,125,150],rwidth=0.95,color=['blue','green'],label=['Male','Female'])
plt.legend()
```

```python
plt.hist([Blood_suger,Blood_suger_women],bins=[80,100,125,150],rwidth=0.95,color=['blue','green'],label=['Male','Female'],orientation='horizontal')
plt.legend()
```

# Pie Chart

```python
marks=[23.5,22.5,30,15,13]
exms=['Mid Term','CT 1','CT 2','Attendence','Performance']
```

```python
plt.pie(marks,labels=exms)
plt.show()
```

```python
plt.pie(marks,labels=exms,radius=2)
plt.show()
```

```python
plt.pie(marks,labels=exms,radius=1.5,autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(marks,labels=exms,radius=1.5,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
plt.pie(marks,labels=exms,radius=1.5,autopct='%0.1f%%',shadow=True,explode=[0.3,0,0,0,0])
plt.show()
```

```python
plt.pie(marks,labels=exms,radius=1.5,autopct='%0.1f%%',shadow=True,startangle=90)
plt.show()
```

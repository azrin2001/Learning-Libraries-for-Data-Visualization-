# Seaborn with Covid Data

```python
df=pd.read_csv('covid_19_dataset.csv')
df
```

```python
df.columns
```

```python
sns.barplot(x=df["continent"],y=df["total_cases"])
plt.xticks(rotation = 60)
plt.show()
```

```python
sns.lineplot(x=df["location"],y=df["total_cases"])
plt.show()
```

```python
sns.scatterplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"])
plt.show()
```

# Color Control

```python
sns.scatterplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],hue=df["continent"])
plt.show()
```

```python
sns.scatterplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],hue=df["continent"],
                hue_order=['Asia','Africa','Europe','South America','North America','Oceania'])
plt.show()
```

```python
hue_colors={'Asia':'black',
           'Africa':'blue',
           'Europe':'yellow',
           'South America':'red',
           'North America':'green',
           'Oceania':'orange'}
sns.scatterplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],hue=df["continent"],palette=hue_colors)
plt.show()
```

```python
sns.countplot(x=df["location"],hue=df["continent"])
plt.show()
```

# relplot()

```python
sns.relplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],kind="scatter")
plt.show()
```

```python
sns.relplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],kind="scatter",col=df["continent"])
plt.show()
```

```python
sns.relplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],kind="scatter",row=df["continent"])
plt.show()
```

```python
sns.relplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],kind="scatter",col=df["continent"],col_wrap=2)
plt.show()
```

```python
sns.relplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],kind="scatter",col=df["continent"],col_wrap=2,col_order=['Asia','Africa','Europe','South America','North America','Oceania'])
plt.show()
```

```python
sns.relplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],kind="scatter",size=df["population"],hue=df["population"])
plt.show()
```

```python
sns.relplot(x=df["total_cases_per_million"],y=df["total_deaths_per_million"],kind="scatter",size=df["population"],hue=df["continent"],style=df["continent"],alpha=0.4)
plt.show()
```

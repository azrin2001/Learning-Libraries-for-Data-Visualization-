```python
import matplotlib.pyplot as plt

x = [2017, 2018, 2019, 2020, 2021]
y = [43, 45, 47, 48, 50]
```

# Lineplot

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle='--')
```

```python
plt.plot(x,y,marker='o',linestyle='--',color='r')
```

```python
plt.plot(x,y,marker='o',linestyle='--',color='r',label='USA')
```

```python
plt.plot(x,y,marker='.',linestyle='--',color='r',label='USA')

plt.xlabel('Years')
plt.ylabel('Population (Millions)')
plt.title('Years vs Population')

plt.legend(loc='lower right')

plt.yticks([42,44,46,48,50])

#Save Plot
plt.savefig('M.png')

plt.show()
```

# Barplot

```python
x = ['USA', 'UK', 'Australia']
y = [40, 50, 33]

plt.bar(x, y,color='g')
plt.xlabel('Countries')
plt.ylabel('Population (Millions)')

plt.show()
```

# Piechart

```python
x = [2017, 2018, 2019, 2020, 2021]
y = [43, 45, 47, 48, 50]

plt.pie(y, labels=x, autopct='%.0f %%')
plt.show()
```

# Histogram

```python
ages = [15, 16, 17, 30, 31, 32, 35]
bins = [15, 20, 25, 30, 35]

plt.hist(ages, bins,color='r', edgecolor='black')
plt.show()
```

# Boxplots

```python
ages = [15, 16, 17, 30, 31, 32, 35]

plt.boxplot(ages)
plt.show()
```
